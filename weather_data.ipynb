```python
import pandas as pd
import numpy as np
import os
```

```python
path = 'weather_data'
files= os.listdir(path)
```

```python
lst = []
index = 1
for name in files:
    print(index)
    lst.append(pd.read_csv(os.path.join(path,name),sep=';'))
    index += 1
df = pd.concat(lst)
df = df.sort_values(by=['Year','Month','Day'],ascending=True).reset_index().drop(['index'],axis = 1)
```
